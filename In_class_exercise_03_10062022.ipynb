Question 2 (20 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction. 

```python
# You code here (Please add comments in the code):
from six.moves import urllib
import pandas as pd

txt_arry = []
url = "https://github.com/sridhar-maddi/Sridhar_INFO5731_FALL2022/blob/main/data.txt"
text_file = urllib.request.urlopen(url)

for line in text_file:
  decoded_data = line.decode("utf-8").replace('\r\n', '')
  if decoded_data: txt_arry.append(decoded_data)
df = pd.DataFrame (txt_arry, columns = ['Raw Text'])
print(df)
```

```python
df['Number of Sentences'] = df['Raw Text'].apply(lambda x: len(str(x).split(".")))
df
```

```python
df['Number of Words'] = df['Raw Text'].apply(lambda x: len(str(x).split(" ")))
df
```

```python
df['Number of Characters'] = df['Raw Text'].str.len()
df
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
  

import nltk
stop = stopwords.words('english')

df['Number of Stopwords'] = df['Raw Text'].apply(lambda x: len([x for x in x.split() if x in stop]))
df
```

```python
df['Lower Case'] = df['Raw Text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df
```

```python
df['Removal of Punctuation'] = df['Lower Case'].str.replace('[^\w\s]','')
df
```

```python
# Graph plot utility
def plot_graphs(history, metric):
  plt.plot(history.history[metric])
  plt.plot(history.history['val_'+metric], '')
  plt.xlabel("Epochs")
  plt.ylabel(metric)
  plt.legend([metric, 'val_'+metric])
```

```python
# Downlading data from tfds and performing test train split
dataset, info = tfds.load('imdb_reviews', with_info=True,
                          as_supervised=True)
train_dataset, test_dataset = dataset['train'], dataset['test']

train_dataset.element_spec
```

```python
# Sample data point
for example, label in train_dataset.take(1):
  print('text: ', example.numpy())
  print('label: ', label.numpy())
```

```python
# Setting buffer and batch sizes
BUFFER_SIZE = 10000
BATCH_SIZE = 64
```

```python
# Shuffling and autotuning data
train_dataset = train_dataset.shuffle(BUFFER_SIZE).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_dataset = test_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
```

```python
# Sample data points after above operations
for example, label in train_dataset.take(1):
  print('texts: ', example.numpy()[:3])
  print()
  print('labels: ', label.numpy()[:3])
```

```python
# Defining vocab size and encoding
VOCAB_SIZE = 1000
encoder = tf.keras.layers.TextVectorization(
    max_tokens=VOCAB_SIZE)
encoder.adapt(train_dataset.map(lambda text, label: text))
```
